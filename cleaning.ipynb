```python
dataset['occupation_type'] = dataset['occupation_type'].map({'Unknown': ''})
dataset['gender'] = dataset['gender'].map({'M':1 , 'F':0})
dataset['owns_car'] = dataset['owns_car'].map({'Y':1 , 'N':0})
dataset['owns_house'] = dataset['owns_house'].map({'Y':1 , 'N':0})
```

```python
dataset.fillna(dataset.select_dtypes(include='number').mean().iloc[0], inplace=True)
dataset.fillna(dataset.select_dtypes(include='object').mode().iloc[0], inplace=True)
```
